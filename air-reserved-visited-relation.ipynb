```python
import pandas as pd
import numpy as np
import matplotlib

path1 = 'reserved.csv'
path2 = 'visited.csv'

# import pandas as pd
train1 = pd.read_csv(path1)
train2 = pd.read_csv(path2)

print (type(train1))
print (type(train2))

import matplotlib.pyplot as plt
rowlimits=13339
reserve = []
visit = []
#count = 0
#reserve_visit = {}
for index1,row1 in train1.iterrows():
    for index2,row2 in train2.iterrows():
        if row1[0]==row2[0] and row1[1]==row2[1]:
            print(index1,index2)
            #reserve_visit.setdefault(row1[2],row2[2])
            reserve.append(row1[2])
            visit.append(row2[2])
plt.scatter(reserve,visit)
plt.legend()
plt.grid(None, 'major', 'both')
plt.title('air-store-reserved-visited-relation'%rowlimits)
plt.xlabel('reserved')
plt.ylabel('visited')
```
